```python
py = df.Language.str.contains('Python').sum()
```

```python
C = df.Language.str.contains('C').sum()
```

```python
Java = df.Language.str.contains('Java').sum() + df.Language.str.contains('js').sum()
```

```python
others = df.Language.count() - (py+C+Java)
```

```python
group_names = ['Python','C','Java','Others']
group_size = [py,C,Java,others]
subgroup_names = []
```

```python
colors = ['#ff9999', '#ffc000', '#8fd9b6', '#d395d0']
wedgeprops={'width': 0.7, 'edgecolor': 'w', 'linewidth': 5}
plt.figure(figsize=(10,10))
plt.pie(group_size, labels=group_names, autopct='%1.1f%%', startangle=260, counterclock=False, colors=colors, wedgeprops=wedgeprops
    ,textprops={'fontsize':12})
plt.legend(group_names,loc=2)
plt.show()
```
